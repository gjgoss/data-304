```python
import pandas as pd
import numpy as npt
from collections import Counter
```

```python
catalog = pd.read_csv('course_catalog_by_major.csv')
pathways = pd.read_csv('course_pathways.csv')
```

```python
catalog
```

```python

ds_required = list(catalog[(catalog['Major'] == 'Data science') & (catalog['Type'] == 'Required')]['Course'])
ds_elective = list(catalog[(catalog['Major'] == 'Data science') & (catalog['Type'] == 'Elective')]['Course'])
ai_required = list(catalog[(catalog['Major'] == 'Artificial intelligence') & (catalog['Type'] == 'Required')]['Course'])
ai_elective = list(catalog[(catalog['Major'] == 'Artificial intelligence') & (catalog['Type'] == 'Elective')]['Course'])
cy_required = list(catalog[(catalog['Major'] == 'Cybersecurity') & (catalog['Type'] == 'Required')]['Course'])
cy_elective = list(catalog[(catalog['Major'] == 'Cybersecurity') & (catalog['Type'] == 'Elective')]['Course'])
```

```python
pathways = pathways[(pathways['Major'] == 'Data science') | (pathways['Major'] == 'Artificial intelligence') | (pathways['Major'] == 'Cybersecurity')]
```

```python
pathways.reset_index(drop=True, inplace=True)
```

```python
pathways
```

```python
paths = []

for path in pathways.CoursePath:
    path = path.split(" -> ")
    paths.append(path)

paths

pathways.CoursePath = paths
```

```python
pathways
```

```python
pathways.loc[0,:]
```

```python
# Retain only the ordered sequence of required courses

course_paths = []

for _, row in pathways.iterrows():
    path = row['CoursePath']
    major = row['Major']
    if major == 'Data science':
        path = [course for course in path if course not in ds_elective]
    if major == 'Artificial intelligence':
        path = [course for course in path if course not in ai_elective]
    if major == 'Cybersecurity':
        path = [course for course in path if course not in cy_elective]

    course_paths.append(path)

pathways.CoursePath = course_paths
```

```python
ind_course_seq = []

for path in pathways.CoursePath:
    course_seq = []
    for n in range(0, len(path)-1):
        seq = path[n] + " -> " + path[n+1]
        course_seq.append(seq)
    ind_course_seq.append(course_seq)

pathways[['t1','t2','t3','t4','t5','t6','t7','t8','t9']] = ind_course_seq
pathways
```

```python
course_sequences = [x for xs in ind_course_seq for x in xs]
course_sequences
counts = dict(Counter(course_sequences))
seq_counts = pd.DataFrame.from_dict(counts, orient='index').reset_index()
seq_counts = seq_counts.rename(columns={'index':'seq',0:'count'})
```

```python
seqs = []

for seq in seq_counts.seq:
    seq = seq.split(" -> ")
    seqs.append(seq)

seq_counts.seq = seqs
```

```python
seq_counts[['course1','course2']] = pd.DataFrame(seq_counts.seq.tolist(), index=seq_counts.index)
seq_counts.drop(columns = 'seq')
seq_counts = seq_counts[['course1','course2','count']]
```
