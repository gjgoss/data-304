```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
system_record = pd.read_csv('system_of_record_timestamps.csv')
```

```python
system_record
```

```python
system_record['Primary_Timestamp'] = pd.to_datetime(system_record['Primary_Timestamp'], format='mixed', utc=True)
system_record
```

```python
system_record.isnull().sum()
```

```python
system_record = system_record.sort_values('Primary_Timestamp')
system_record
```

```python
system_record_oct_nov = system_record[[(mon == 11) | (mon == 10) for mon in [date.month for date in system_record.Primary_Timestamp]]]
system_record_oct_nov
```

```python
system_record.Primary_Timestamp[0].month
```

```python
plt.plot(system_record_oct_nov.Primary_Timestamp, system_record_oct_nov.Event_Count)
plt.xlabel('Time')
plt.ylabel('Events')
plt.title('System Events')
plt.savefig('assignment_4_graph')
```
